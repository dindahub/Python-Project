### Correlation Data

Dari sini kita bisa melihat variabel Unit Cost sampai Revenue memiliki korelasi yang kuat daripada Day sampai order quantity, lalu bagaimana dengan variabel kategori? Kita akan merubah variabel kategorikal mejadi angka

```python
sns.heatmap(dt.corr(), annot=True, cmap="Blues", fmt='.1g')
```

```python
from sklearn.preprocessing import LabelEncoder

lb = LabelEncoder()
dt['Customer_Gender_Index'] = lb.fit_transform(dt['Customer_Gender'])
dt['Age_Group_Index'] = lb.fit_transform(dt['Age_Group'])
dt.head()
```

Dapat dilihat bahwa **Age Index** dan ***Gender Index*** tidak memiliki korelasi yang signifikan

```python
sns.heatmap(dt.corr(), annot=True, cmap="Blues", fmt='.1g')
```

Dari visualisasi pairplot kita bisa melihat variabel yang berkoleralasi dan memiliki pola. Dari visual ini juga kita bisa melihat outlier dari beberapa variabel. Untuk itu kita memerlukan pembersihan data dengan beberapa cara

```python
sns.pairplot(dt)
```

### Data Cleaning
* Clean Outlier
* Drop null value
* Disini kita tidak melakukan drop duplicat karena data tidak memilki **primary key**

```python
dt.dropna(inplace=True)
dt.shape
```

Dataset ini ternyata tidak memiliki null value

Dari grafik dapat dilihat daaset memiliki dua outlier sehingga kita harus menghapus atau memfilternya

```python
sns.scatterplot(x='Cost', y='Revenue', data=dt)
```

```python
out= np.where(dt['Cost'].values>30000)
```

```python
arr = np.array(out).flatten()
arr
```

```python
dt.drop(index=arr, inplace=True)
dt.shape
```

```python
sns.scatterplot(x='Cost', y='Revenue', data=dt)
```

### Filter Data

```python
sns.scatterplot(x='Unit_Cost', y='Unit_Price', data=dt)
```

```python
dt_filter = dt[dt['Unit_Price']<2500]
dt_filter.shape
```

```python
sns.scatterplot(x='Year', y='Order_Quantity', data=dt_filter)
```

```python
dt_filter = dt_filter[(dt_filter['Year']>2012) & (dt_filter['Customer_Age']<75)]
dt_filter.shape
```

Setelah dilakukan proses filtering data, jumlah data menjadi **104636**

### Analysis Data

**Rata-rata, min, max**

```python
dt_filter.groupby(['Age_Group']).agg({'Profit': ['mean', 'max','min'], 'Revenue': ['mean', 'max','min'], 'Order_Quantity': ['mean', 'max','min']})
```

Kita analisis Age Group berdasarkan profit, revenue, dan order quantity menggunakan groupby. Hasil menunjukkan Umur 64 keatas memiliki profit yang paling rendah tapi mempunyai rata-rata jumlah order yang tinggi.

```python
dt_filter.groupby(['Customer_Gender']).agg({'Profit': ['mean', 'max','min'], 'Revenue': ['mean', 'max','min'], 'Order_Quantity': ['mean', 'max','min']})
```

Jika kita analisis berdasarkan customer gender, perempuan lebih memiliki profiit dan revenue yang tinggi dibandingkan dengan laki-laki. Namun Rata-rata jumlah order lebih tinggi laki-laki.

```python
dt_filter.groupby(['Country']).agg({'Profit': ['mean', 'max','min'], 'Revenue': ['mean', 'max','min'], 'Order_Quantity': ['mean', 'max','min']}).sort_values(by=('Profit','mean'), ascending=False)
```

Jika kita analisis berdasarkan country, Inggris dan Jerman lebih memiliki profiit dan revenue yang tinggi dibandingkan dengan negara lain. Namun Rata-rata jumlah order lebih tinggi kanada. Ini menarik, karena customer di Amerika Serikat lebih tinggi dari Inggris, namun profit dan pendapatannya lebih rendah dari Inggris dan Jerman

```python
dt_filter.groupby(['Sub_Category']).agg({'Profit': ['mean', 'max','min'], 'Revenue': ['mean', 'max','min'], 'Order_Quantity': ['mean', 'max','min']}).reset_index().sort_values(by=('Profit','mean'), ascending=False)
```

**Time**

```python
months = ["January", "February", "March", "April", "May", "June", 
          "July", "August", "September", "October", "November", "December"]
dtmp = dt_filter.pivot_table(values='Profit', index=['Month'], columns=['Sub_Category'], aggfunc=np.mean).sort_values('Month', key = lambda x : pd.Categorical(x, categories=months, ordered=True)).T
```

```python
sns.heatmap(dtmp, cmap="Blues", annot=True, fmt=".0f")
```

```python
dtmr = dt_filter.pivot_table(values='Revenue', index=['Month'], columns=['Sub_Category'], aggfunc=np.mean).sort_values('Month', key = lambda x : pd.Categorical(x, categories=months, ordered=True)).T
sns.heatmap(dtmr, cmap="Blues", annot=False, fmt=".0f")
```

Jika kita analisis dari bulan, profit dan revenue dari produk tidak mengalami peningkatan yang signifikan. Produk sepeda mempunyai penjualan yang lebih tinggi dibandingkan dengan produk lain

```python
dtmy = dt_filter.pivot_table(values='Profit', index=['Month'], columns=['Year'], aggfunc=np.mean).sort_values('Month', key = lambda x : pd.Categorical(x, categories=months, ordered=True)).T
sns.heatmap(dtmy, cmap="Blues", annot=False, fmt=".0f")
```

Dari sini baru bisa kita lihat bahwa profit tahun 2013 dan 2015 memiliki nilai yang tinggi terutama pada bulan antara januari sampai Juni dan menurun di bulan juli

```python
sns.set()
```

```python
sns.relplot(data=dtmy.T, kind="line", height=9)
```
